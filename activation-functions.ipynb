### Sigmoid
It is one of the most widely used non-linear activation function,it transforms the values between the range 0 and 1.
- **Math Formula**: - $$ \frac {1} {(1 + e^{-x})} $$
- **PyTorch**:- `nn.Sigmoid()`
- **TensorFlow**:- `tf.keras.activations.sigmoid()`

you can read more about them here - https://deepai.org/machine-learning-glossary-and-terms/sigmoid-function

```python
# Creating the function from Scratch
def Sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
# Plotting
fig = make_subplots(rows=1,cols=3,subplot_titles=["Regular","Sigmoid Function","PyTorch Sigmoid Function"],x_title="Data")
sig = nn.Sigmoid()
fig.add_scatter(y= data,row=1,col=1)
fig.add_scatter(x = data,y= Sigmoid(data),row=1,col=2)
fig.add_scatter(x = data,y= sig(torch.from_numpy(data)),row=1,col=3)
```

### Rectified Linear Unit (ReLU) 
ReLU Activation Function will output the input directly if it is positive, otherwise, it will output zero
$$ max(0.0 , x) $$
- PyTorch:- `nn.ReLU`
- TensorFlow:- `tf.keras.activations.relu`

```python
# Creating ReLU using Code
def ReLU(x):
 return np.maximum(0.0, x)
```

```python
# Plotting
relu = nn.ReLU()
fig = make_subplots(rows=1,cols=3,subplot_titles=["Regular","ReLU Function","PyTorch ReLU Function"],x_title="Data")
fig.add_scatter(y= data,row=1,col=1)
fig.add_scatter(x = data,y= ReLU(data),row=1,col=2)
fig.add_scatter(x = data,y= relu(torch.from_numpy(data)),row=1,col=3)
```

### Tanh
* https://www.mathsisfun.com/definitions/tanh.html
$$ Tanh =  \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}  $$

```python
# Creating Leaky ReLU using Code
def Tanh(x):
 return ((np.exp(x))-(np.exp(-x)))/((np.exp(x))+(np.exp(-x)))
```

```python
# Plotting
tanh = nn.Tanh()
fig = make_subplots(rows=1,cols=3,subplot_titles=["Regular","Tanh Function","PyTorch Tanh Function"],x_title="Data")
fig.add_scatter(y= data,row=1,col=1)
fig.add_scatter(x = data,y= Tanh(data),row=1,col=2)
fig.add_scatter(x = data,y= tanh(torch.from_numpy(data)),row=1,col=3)
```

### Parameterised ReLU (PReLU)


```python
# Creating Leaky ReLU using Code
def PReLU(x,alpha):
    return np.maximum(0,x) + alpha * np.minimum(0,x) 
```

# Softmax
The softmax function is a function that turns a vector of K real values into a vector of K real values lie in the range [0,1] and sum to 1.
$$\sigma(z)_i = \frac{e^{z_i}}{\sum^k_{j=1} e^{z_j}} $$ OR $$ Softmax(x_i) = \frac{exp(x_i)}{\sum_j exp(x_j)} $$

- https://deepai.org/machine-learning-glossary-and-terms/softmax-layer
- https://pytorch.org/docs/stable/generated/torch.nn.Softmax.html

```python
def Softmax(x):
    return (np.exp(x))/(np.sum(np.exp(x)))

soft = nn.Softmax(dim = 1)
inp = torch.randn(2,3)
# soft(inp)
# Softmax(inp)
data2=torch.from_numpy(data).reshape(-1,1)
```

# Binary Step Function
Binary step function is one of the most common and simplest activation function in neural networks, produces 1 (or true) when input passes a threshold limit whereas it produces 0 (or false) when input does not pass threshold.
- https://iq.opengenus.org/binary-step-function/


```python
def BinaryStep(x):
    return (x>0).astype(int)
```

```python
sign = torch.sign(torch.from_numpy(data))
binary = torch.relu(sign)
binary
```

```python
binary = torch.relu(torch.sign(torch.from_numpy(data)))
fig = make_subplots(rows=1,cols=3,subplot_titles=["Regular","Binary Step Function","PyTorch Binary Step Function"],x_title="Data")
fig.add_scatter(y= data,row=1,col=1)
fig.add_scatter(x = data,y= BinaryStep(data),row=1,col=2)
fig.add_scatter(x = data,y= binary,row=1,col=3)
```

# Sigmoid-Weighted Linear Units (SiLU) also Known as Swish Function

$$silu(x) = x * \sigma (x) $$ 
> where $\sigma (x)$ is the logistic sigmoind


```python
def SiLU(x):
    return (x*Sigmoid(x))
```

```python
silu = torch.from_numpy(data) * torch.sigmoid(torch.from_numpy(data))
fig = make_subplots(rows=1,cols=3,subplot_titles=["Regular","Binary Step Function","PyTorch Binary Step Function"],x_title="Data")
fig.add_scatter(y= data,row=1,col=1)
fig.add_scatter(x = data,y= SiLU(data),row=1,col=2)
fig.add_scatter(x = data,y= silu,row=1,col=3)
```
